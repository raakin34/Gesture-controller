```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Function to perform polynomial regression and plot for each sample
def polynomial_trend_for_each_sample(gesture_data, degree=3):
    num_samples = gesture_data.shape[0]

    # Iterate through each sample and fit a polynomial regression for its tilt values
    for i in range(num_samples):
        tilt_values = gesture_data[i, :, 2]  # Extract the tilt values (third column) for this sample
        X = np.arange(len(tilt_values)).reshape(-1, 1)  # X is the index (time/frame)
        y = tilt_values  # y is the tilt value

        # Polynomial transformation
        poly = PolynomialFeatures(degree=degree)
        X_poly = poly.fit_transform(X)

        # Fit polynomial regression
        poly_regressor = LinearRegression()
        poly_regressor.fit(X_poly, y)

        # Predict the tilt values using the polynomial model
        y_poly_pred = poly_regressor.predict(X_poly)

        # Plotting the tilt values and the polynomial trend for this sample
        plt.figure(figsize=(10, 6))
        plt.plot(tilt_values, marker='o', linestyle='-', color='b', label=f'Tilt Values - Sample {i+1}')
        plt.plot(X, y_poly_pred, color='r', linestyle='--', label=f'Polynomial Trend (Degree {degree})')
        plt.title(f'Trend in Tilt Values for Sample {i+1} (Degree {degree})')
        plt.xlabel('Index (Time/Frame)')
        plt.ylabel('Tilt Value')
        plt.grid(True)
        plt.legend()
        plt.show()

        # Optionally print the polynomial coefficients for each sample's trend
        print(f"Sample {i+1} Polynomial Coefficients: ", poly_regressor.coef_)
        print(f"Sample {i+1} Polynomial Intercept: ", poly_regressor.intercept_)

# Call the function with gesture data and degree 3 for polynomial fitting
polynomial_trend_for_each_sample(gesture_data, degree=3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error, r2_score

# Function to perform machine learning (polynomial regression) and validate new data
def polynomial_trend_with_ml_and_validation(gesture_data, new_data, degree=3):
    num_samples = gesture_data.shape[0]

    # Iterate through each sample and fit a polynomial regression model for its tilt values
    for i in range(num_samples):
        tilt_values = gesture_data[i, :, 2]  # Extract the tilt values (third column) for this sample
        X = np.arange(len(tilt_values)).reshape(-1, 1)  # X is the index (time/frame)
        y = tilt_values  # y is the tilt value

        # Create a machine learning pipeline with PolynomialFeatures and LinearRegression
        model = make_pipeline(PolynomialFeatures(degree), LinearRegression())

        # Fit the model
        model.fit(X, y)

        # Predict the tilt values using the fitted model
        y_pred = model.predict(X)

        # Plotting the tilt values and the polynomial trend for this sample
        plt.figure(figsize=(10, 6))
        plt.plot(tilt_values, marker='o', linestyle='-', color='b', label=f'Tilt Values - Sample {i+1}')
        plt.plot(X, y_pred, color='r', linestyle='--', label=f'ML Polynomial Trend (Degree {degree})')
        plt.title(f'Machine Learning Trend in Tilt Values for Sample {i+1} (Degree {degree})')
        plt.xlabel('Index (Time/Frame)')
        plt.ylabel('Tilt Value')
        plt.grid(True)
        plt.legend()
        plt.show()

        # Optionally print the model coefficients and intercept
        print(f"Sample {i+1} Polynomial Coefficients: ", model.named_steps['linearregression'].coef_)
        print(f"Sample {i+1} Polynomial Intercept: ", model.named_steps['linearregression'].intercept_)

        # Validate using the new data: Assuming new_data is a list of tilt values
        if new_data.shape[0] == len(tilt_values):
            X_new = np.arange(len(new_data)).reshape(-1, 1)
            y_new_pred = model.predict(X_new)

            # Evaluate the error between the new data and predicted values
            mse = mean_squared_error(new_data, y_new_pred)
            r2 = r2_score(new_data, y_new_pred)

            print(f"Validation for Sample {i+1}:")
            print(f"Mean Squared Error (MSE): {mse}")
            print(f"R-squared (R2): {r2}")

            # You can set a threshold to determine if the new data fits well
            if mse < 600:  # Adjust this threshold based on your tolerance
                print(f"New data fits well with the trend for Sample {i+1}.")
            else:
                print(f"New data does NOT fit well with the trend for Sample {i+1}.")
        else:
            print(f"Warning: The new data length doesn't match the sample {i+1} length. Skipping validation.")

# New tilt values (replacing your previous set)
new_data = np.array([
    -15.36, 11.21, 17.43, 35.38, 39.05, 45.26, 45.22, 34.84, 28.69, -10.3,
    -16.27, -51.99, -61.62, -82.09, -86.69, -99.32, -99.9, -87.74, -82.71,
    -52.19, -43.25, 0.0, 0.0, 0.0, 0.0
])







# Call the function with gesture data and degree 3 for polynomial fitting
polynomial_trend_with_ml_and_validation(gesture_data, new_data, degree=3)
```
